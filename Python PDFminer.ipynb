```python
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdfparser import PDFParser
```

```python
from pdfminer.pdfparser import PDFParser
```

```python
document = open('c:/data/pdfminer-docs.pdf', 'rb')
#리소스 매니저 생성
rsrcmgr = PDFResourceManager()
# 분석 파라미터 설정
laparams = LAParams()

device = PDFPageAggregator(rsrcmgr, laparams=laparams)
interpreter = PDFPageInterpreter(rsrcmgr, device)

for page in PDFPage.get_pages(document):
    interpreter.process_page(page)
    layout = device.get_result()
    for element in layout:
        if isinstance(element, LTTextBox):
            print(element.get_text())
```

```python
fp = open('c:/data/pdfminer-docs.pdf', 'rb')
parser = PDFParser(fp)
document = PDFDocument(parser)
outlines = document.get_outlines()
for (level, title, dest, a, se) in outlines:
    print(level, title)
```

####  Linked open data 수집

```python
import SPARQLWrapper
```

```python
from SPARQLWrapper import SPARQLWrapper
```

```python
sparql = SPARQLWrapper('http://ko.dbpedia.org/sparql')

sparql.setQuery('''
SELECT * WHERE{
    ?s rdf:type dbpedia-owl:Musewm .
    ?s prop-ko: 소재지 ?address .
} order by ?s
''')

sparql.setReturnFormat('json')

res = sparql.query().convert()
for result in res['results']['bindings']:
    print(result['s']['value'], result['address']['value'])
```

#### 구글에서 사용하기

```python
import robobrowser
```

```python
from robobrowser import RoboBrowser

browser = RoboBrowser(parser='html.parser')
browser.open('https://www.google.co.kr/')

form = browser.get_form()
form['q'] = 'python'
browser.submit_form(form, submit=form.submit_fields['btnG'])

for a in browser.select('a'):
    print(a.text)
    print(a.get('href'))
    print('======')
```
