람다 함수의 장점  
메모리 절약 + 가독성 향상  
일반 함수는 객체를 만들고, 재사용을 위해 함수 이름(메모리)를 할당한다.  
람다 함수는 한 번 사용하고 다음 줄로 넘어가면 메모리(힙(heap))에서 사라진다.

#### lambda를 활용해서 계산기 프로그램 만들기

1. 화면 구성하기  
2. 버튼 활성화 하기

```python
from tkinter import *

window= Tk() #윈도우 창 부르기
window.title("My Calculator") #타이틀 설정
display= Entry(window, width=33, bg="yellow") #크기 설정, 배경색
display.grid(row=0, column=0, columnspan=5)

button_list = [
    '7','8','9','/','C',
    '4','5','6','*','**', 
    '1','2','3','-','%', 
    '0','.','=','+','←'
]

def click(key):# 버튼 활성화를 위한 이벤트 연동하기 함수 만들기
    if key == '=':
        result = eval(display.get())
        s= str(result)
        display.insert(END,'='+s)
    elif key == 'C':
        display.delete(0,END)
    elif key == '←':
        display.delete(len(display.get())-1, END)
    else:
        display.insert(END, key)


col_index=0
rw_index=1
for button_text in button_list:
    Button(window, text=button_text, width=5, command=lambda x = button_text:click(x)).grid(row=rw_index    ,column=col_index)
    col_index +=1
    if col_index%5==0:
        col_index=0
        rw_index+=1
        
window.mainloop() #항상 코드 가장 아래쪽에 위치한다.
```

#### 메뉴 위젯 만들기

```python
from tkinter import *
from tkinter.filedialog import * #파일 다이얼로그 박스 파일 불러오기 저장하기 등
import tkinter.messagebox
import tkinter.simpledialog
import tempmodule as t # 온도변화프로그램 함수 임포트하기
import calmodule as c # 계산기프로그램 함수 임포트하기
window=Tk()
memo_menu = Menu(window) #Menu의 위젯()으로 mymenu를 만들기
window.title("MyMenu")
window.config(menu=memo_menu)

text=Text(window, height=30, width=80)
text.pack()

def save():
    file= asksaveasfile(parent=window, mode='w')
    print(file)
    if file != None:
        lines = text.get('0.0', END+'-1c') # 0행 0열부터 끝가지 get해서 lines에 저장, '-1c'는 마지막 빈줄 넣지 말라는 의미
        file.write(lines)
        file.close()

def open():
    file = askopenfile(parent= window, mode='r')
    if file != None:
        lines = file.read()
        text.insert('0.0', lines)
        file.close()

def destroy():
    window.destroy()

def msginfo():
    tkinter.messagebox.showinfo('showinfo', '이것은 알림 메시지입니다.')
    
def exit():
    if tkinter.messagebox.askokcancel('종료', '메모장을 종료하시겠습니까?'):
        window.destroy()


def new():
    yn=tkinter.messagebox.askyesnocancel('새로만들기','새로만드시겠습니까?')
    if yn==True:
        save()
        text.delete('0.0',END)
    elif yn==False:
        text.delete('0.0',END)
    else:
        pass
    
def summ():
    num=tkinter.simpledialog.askinteger('합구하기', '숫자를 입력하세요')
    num=(num*(num+1))//2
    tkinter.messagebox.showinfo('showinfo', num)
    

filemenu=Menu(memo_menu)

memo_menu.add_cascade(label='파일', menu=filemenu)
filemenu.add_command(label='새로만들기', command=new)
filemenu.add_command(label='열기', command=open)
filemenu.add_command(label='저장하기', command=save)
filemenu.add_command(label='종료', command=exit)

programmenu = Menu(memo_menu)
memo_menu.add_cascade(label = '프로그램', menu=programmenu)
programmenu.add_command(label='합구하기프로그램', command=summ)
programmenu.add_command(label='온도변환프로그램', command=t.tempmodule) #불러온 모듈에서 필요한 함수를 부른다 (모듈에는 함수 3개가 존재함)
programmenu.add_command(label='계산기프로그램', command=c.calmodule)
helpmenu=Menu(memo_menu)
showmenu=Menu(window)
memo_menu.add_cascade(label='도움말', menu=helpmenu)
helpmenu.add_cascade(label='프로그램정보', menu=showmenu)
showmenu.add_command(label='알림', command=msginfo)


window.mainloop()
```

#### 아래 3개의 코드는 서브 메뉴를 만들기를 실습해본 코드

```python
from tkinter import* #모듈 -> 파일안 함수는 모두 import 가능
from tkinter.filedialog import* # filedialog는 Sub 모듈이므로 가져오지 못한다.
import tkinter.messagebox


w = Tk()
w.title('나의 메모장')
text = Text(w, height=30, width=80)
text.pack()

def save():
    file = asksaveasfile(parent=w, mode='w') # w 쓰기모드
    if file != None:
        lines = text.get('0.0', END+'-1c')
        # 0행 0열부터 끝까지 get해서 lines에 저장, '-1c'는 마지막 빈줄 넣지 말란 의미
        file.write(lines)
        print(file)
        file.close()

def open():
    file = askopenfile(parent=w, mode='r') # w 쓰기모드
    if file != None:
        lines = file.read()
        text.insert('0.0',lines)
        file.close()
               
def exit():
    w.destroy()

def info():
    tkinter.messagebox.showinfo("showinfo", "이것은 내가 만든 정보창입니다.")
    

mymenu = Menu(w) # Menu위젯()으로 mymenu를 만듦
infomenu = Menu(w)

w.config(menu=mymenu) # mymenu를 윈도우에 연결해 줌
filemenu=Menu(mymenu) # mymenu의 메뉴로 filemenu를 만듦
mymenu.add_cascade(label='파일', menu=filemenu) # filemenu 에 '파일'을 드롭다운방식으로 추가
filemenu.add_cascade(label='열기', command=open) # "파일"아래 서브메뉴 "열기"를 추가함.
filemenu.add_cascade(label='저장', command=save) # "파일"아래 서브메뉴 "저장"를 추가함.
filemenu.add_cascade(label='종료', command=exit) # "파일"아래 서브메뉴 "종료"를 추가함.

helpmenu = Menu(mymenu)
mymenu.add_cascade(label='도움말', menu=helpmenu)
helpmenu.add_cascade(label='도움말2', menu=infomenu)
infomenu.add_command(label='프로그램 정보', command=info)

w.mainloop()
```
