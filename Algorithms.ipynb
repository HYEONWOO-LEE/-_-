### 동명이인 찾기 알고리즘

```python
name=['spot','spot','tom','jerry','mike','tom', 'jerry']
s=set()
def name_n(name):
    for i in name:
        name.pop(name.index(i))
        for j in range(0,len(name)):
            if i==name[j]:
                s.add(i)
    return s
```

```python
name_n(name)
```

```python
name
```

위 알고리즘은 리스트를 건드리기 때문에 실패

```python
name=['spot','spot','tom','jerry','mike','tom', 'jerry']
s=set()
def namen_(name):
    for i in range(0,len(name)):
        for j in range(i+1, len(name)):
            if name[i]==name[j]:
                s.add(name[i])
    return s
```

s를 함수 밖에 정의해서 set 집합 내용이 누적이 되었음

```python
namen_(name)
```

```python
name
```

```python
name2=['a','b','c','d','e','a','a','c','h','k','h']
namen_(name2)
```

재귀 호출 프로그램이 정상적으로 작동하려면 '종료 조건'이 필요  
특정 조건이 되면 더이상 자기 자신을 호출하지 않도록 설계되어야 함  
그렇지 않으면 위와 같이 에러 발생  
재귀 호출 함수가 계산 결과를 돌려줄 때는 return을 사용해서 결과값을 돌려줌

팩토리얼 구하기

```python
def fct(n):
    f=1
    if n==1:
        return 1
    else:
        f*=n*(n-1)
    fct(n-1)
    return f
```

```python
fct(4)
```

나의 답안 (오답)

```python
def fct1(n):
    if n <=1:
        return 1
    return n*fct(n-1)
```

재귀함수를 사용해 1부터 n까지 합 구하기

```python
def sumn(n):
    if n<=1:
        return 1
    return n+sumn(n-1)
```

재귀 함수를 사용하여 숫자 n개 중에서 최댓값 찾기

```python
def fnd_mx(a,n): # 리스트 a의 앞부분 n개 중 최댓값 구하기
    if n==1:
        return a[0] # 종료 조건 n이 1이면 리스트 안에 숫자의 개수가 1개라는 뜻이므로 리스트a의 첫 값을 리턴해주면 본인이 나온다
    mx_n_1=fnd_mx(a, n-1) # 재귀함수를 불러 n-1개 중 비교
    if mx_n_1 >a[n-1]:
        return mx_n_1
    else:
        return a[n-1]
```

```python
lst=[1,54,21,56,2,20,18,95,13,102,15]
fnd_mx(lst,len(lst))
```

#### 재귀 함수 원리

```python
i=0
def sum(n):
    global i
    i+=1
    print(i, '번째','매개변수',n)
    if n==1:
        return 1
    sum(n-1)
    i-=1
    print(i, '번째', '매개변수',n)
    
print(sum(10))
```
